```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv(f'{os.getcwd()}/../data/processed.csv')

# pclass = titanic.pclass.name
# survived = titanic.survived.name
# sex = titanic.sex.name
```

```python
plt.figure(figsize=(1.5, 3))
sns.countplot(x=titanic.survived) \
    .set_xticks(ticks=[0, 1], labels=['dead', 'alive'])
plt.show()
```

```python
plt.figure(figsize=(3, 3))
sns.countplot(x=titanic.sex, hue=titanic.survived)
plt.show()
```
